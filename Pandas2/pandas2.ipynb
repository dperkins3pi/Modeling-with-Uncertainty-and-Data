# Problem 1

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    data = pd.read_csv("crime_data.csv", index_col=0)  # Load in data
    # data.drop(["Population"], axis=1).plot()
    
    # Property vs. Violent crimes
    data.plot(y=["Total", "Violent", "Property"], title="Property vs. violent crimes throughout the years", ylabel="Number of crimes")

    # Scatter plot    
    data.drop(["Population"], axis=1).plot(kind="scatter", x="Vehicle Theft", y="Forcible Rape", title="Correlation between vehicle theft and rape")
    
    # Histogram of murder and robbery
    axes = data[["Murder", "Robbery"]].hist(grid=False,bins=8,sharey=True)
    axes[0,0].set_xlabel("Number of crimes")
    axes[0,0].set_xlabel("Quantity of years")
    axes[0,1].set_xlabel("Number of crimes")
    axes[0,1].set_xlabel("Quantity of years")
    plt.suptitle('Quantities of crime throughout the years')
```

```python
prob1()
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft 
            as box plots
        2. The distributions of Vehicle Thefts against Robberies as
            a hexbin plot.
    All plots should be labelled and easy to read.
    """
    data = pd.read_csv("crime_data.csv", index_col=0)  # Load in data
    data.plot(kind="box", y=["Burglary", "Violent", "Vehicle Theft"], title="Distributions of Burglary, Violent, and Vehicle Theft", ylabel="Quantity")  # Box plots
    data.plot(kind="hexbin", x="Vehicle Theft", y="Robbery", gridsize=15, title="Vehicle Thefts against Robberies", sharex=False)  # Hexplots
```
